```python
import pandas as pd
```

```python
df=pd.read_excel(r"C:\Users\tcrac\Downloads\Online Retail.xlsx")
df
```

```python
df.head()
```

```python
df.tail()
```

```python
#data cleaning
# Drop duplicates
df.drop_duplicates(inplace=True)
```

```python
df.isnull()
```

```python
# Remove cancelled orders (IvoiceNo starting with 'C')
df = df[~df['InvoiceNo'].astype(str).str.startswith('C')]
```

```python
# Remove rows with missing customer identifiers
df = df.dropna(subset=['CustomerID'])
```

```python
# Remove negative or zero quantity
df = df[df['Quantity'] > 0]
```

```python
df.head()
```

```python
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df['Month'] = df['InvoiceDate'].dt.month
df['DayOfWeek'] = df['InvoiceDate'].dt.day_name()
df['Hour'] = df['InvoiceDate'].dt.hour
```

```python
df.head()
```

```python
monthly_sales = df.groupby('Month')['TotalPrice'].sum()
monthly_sales.plot(kind='bar', title='Monthly Revenue', color='skyblue')
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.show()
```

### 📈 Business Insight: Monthly Sales Trends

- The highest sales occurred in *November*, suggesting a potential seasonal trend (holiday sales, bulk ordering).
- Months like *February and August* saw lower revenue. The company could explore promotions during these months to balance sales.

 Recommendation: Consider ramping up inventory and marketing before November to capture high demand efficiently.

```python
top_products = df.groupby('Description')['Quantity'].sum().sort_values(ascending=False).head(10)
top_products.plot(kind='barh', title='Top Selling Products', color='orange')
plt.xlabel("Total Quantity Sold")
plt.gca().invert_yaxis()
plt.show()
```

# Insight: Best-Selling Products
- The most sold items are primarily low-cost, high-demand products.
- Consider bundling these with slower-moving items to increase total cart value.

```python
country_sales = df.groupby('Country')['TotalPrice'].sum().sort_values(ascending=False).head(10)
country_sales.plot(kind='bar', title='Revenue by Country')
plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.show()
```

# Insight: Top Countries by Revenue
- The UK is the dominant market (likely HQ-based), followed by Netherlands and Germany.
- Emerging markets like France and EIRE show strong potential.
- Recommendation: Consider expanding warehouse/distribution in top 5 countries.

```python
top_customers = df.groupby('CustomerID')['TotalPrice'].sum().sort_values(ascending=False).head(10)
top_customers.plot(kind='bar', title='Top 10 Customers by Revenue', figsize=(10,6))
plt.ylabel("Revenue")
plt.xlabel("Customer ID")
plt.show()
```

# Insight: High-Value Customers
A small set of customers are responsible for a large share of revenue.
Consider personalized offers, loyalty programs, and retention strategies.

- Peak month is November → prep for holiday sales.
